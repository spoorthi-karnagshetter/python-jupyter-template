### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

try:
    # Data
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    colors = ['blue', 'green', 'red', 'purple']

    # Ensure labels and sizes are not empty and have the same length
    if not labels or not sizes:
        raise ValueError("Labels or sizes list is empty.")
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes lists must have the same length.")

    # Ensure sizes contain positive values
    if any(s <= 0 for s in sizes):
        raise ValueError("Sizes must contain positive values.")

    # Create pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)

    # Add title
    plt.title('Programming Language Popularity')

    # Show plot
    plt.show()

except ValueError as ve:
    print(f"ValueError: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
