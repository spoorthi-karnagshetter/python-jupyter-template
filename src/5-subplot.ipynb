### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

try:
    # Data for plots
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]

    # Ensure data lists are not empty and have correct lengths
    if not x or not y:
        raise ValueError("X or Y data list is empty.")
    if len(x) != len(y):
        raise ValueError("X and Y lists must have the same length.")
    if not categories or not values:
        raise ValueError("Categories or values list is empty.")
    if len(categories) != len(values):
        raise ValueError("Categories and values lists must have the same length.")
    if not labels or not sizes:
        raise ValueError("Labels or sizes list is empty.")
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes lists must have the same length.")

    # Create a 2x2 subplot layout
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    # Line Chart (First Subplot)
    axes[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
    axes[0, 0].set_title('Line Chart')
    axes[0, 0].set_xlabel('X values')
    axes[0, 0].set_ylabel('Y values')

    # Bar Chart (Second Subplot)
    axes[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
    axes[0, 1].set_title('Bar Chart')
    axes[0, 1].set_xlabel('Categories')
    axes[0, 1].set_ylabel('Values')

    # Scatter Plot (Third Subplot)
    axes[1, 0].scatter(x, y, color='blue')
    axes[1, 0].set_title('Scatter Plot')
    axes[1, 0].set_xlabel('X values')
    axes[1, 0].set_ylabel('Y values')

    # Pie Chart (Fourth Subplot)
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'purple'])
    axes[1, 1].set_title('Pie Chart')

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()

except ValueError as ve:
    print(f"ValueError: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
