### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

try:
    # Data
    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    colors = ['blue', 'green', 'red', 'purple']

    # Ensure categories and values are not empty and have the same length
    if not categories or not values:
        raise ValueError("Categories or values list is empty.")
    if len(categories) != len(values):
        raise ValueError("Categories and values lists must have the same length.")

    # Create bar plot
    plt.figure(figsize=(6, 4))
    plt.bar(categories, values, color=colors)

    # Add title and labels
    plt.title('Bar Plot of Categories vs Values')
    plt.xlabel('Categories')
    plt.ylabel('Values')

    # Show plot
    plt.show()

except ValueError as ve:
    print(f"ValueError: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
