### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

try:
    # Data
    x = [1, 2, 3, 4, 5]
    y = [2, 4, 6, 8, 10]

    # Ensure data lists are not empty and have the same length
    if not x or not y:
        raise ValueError("X or Y data list is empty.")
    if len(x) != len(y):
        raise ValueError("X and Y lists must have the same length.")

    # Create scatter plot
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, color='blue', label='Data Points')

    # Annotate each point
    for i in range(len(x)):
        plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')

    # Add title and labels
    plt.title('Scatter Plot with Annotations')
    plt.xlabel('X values')
    plt.ylabel('Y values')
    plt.legend()

    # Show plot
    plt.show()

except ValueError as ve:
    print(f"ValueError: {ve}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
